## base基类

这部分介绍的基类是框架中智能体之间交流以及工作流形成必不可少的部分，会频繁出现在Workflow实现的各个流程中。

共有以下6个部分：

1. TaskRunner

2. Response

3. TaskResult

4. Team

5. Handoff

6. TerminationCondition

由于这部分的基类基本都是声名，是用来搭建底层框架的，后续的很多类都是继承上述基类进行补充或重写，所以我们看一遍源代码了解就行。

### Team

| 方法名                       | 类型          | 返回类型                             | 用途                   |             |                      |
| ------------------------- | ----------- | -------------------------------- | -------------------- | ----------- | -------------------- |
| `on_messages(...)`        | `async def` | `Response`                       | 处理一组传入消息并返回整体响应（非流式） |             |                      |
| `on_messages_stream(...)` | `def`       | \`AsyncGenerator\[BaseAgentEvent | BaseChatMessage      | Response]\` | 流式响应接口，逐步返回中间消息和最终响应 |
| `on_reset(...)`           | `async def` | `None`                           | 重置智能体到初始化状态          |             |                      |
| `on_pause(...)`           | `async def` | `None`                           | 智能体暂停时被调用，可能在处理消息过程中 |             |                      |
| `on_resume(...)`          | `async def` | `None`                           | 智能体恢复运行时调用           |             |                      |
| `save_state()`            | `async def` | `Mapping[str, Any]`              | 保存智能体当前状态            |             |                      |
| `load_state(state)`       | `async def` | `None`                           | 加载之前保存的状态            |             |                      |
| `close()`                 | `async def` | `None`                           | 清理和释放资源              |             |                      |
